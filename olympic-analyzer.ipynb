```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("athlete_events.csv")
noc_df = pd.read_csv("noc_regions.csv")
```

```python
df.head()
```

```python
df = df[df['Season'] == 'Summer']
```

```python
df = df.merge(noc_df, on='NOC', how='left')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = pd.concat([df, pd.get_dummies(df['Medal'])], axis=1)
```

```python
medal_taily = df.drop_duplicates(subset=['Team','NOC','Games','Year','City','Sport','Event','Medal'])
```

```python
medal_taily.groupby('NOC').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending=False).reset_index()
```
